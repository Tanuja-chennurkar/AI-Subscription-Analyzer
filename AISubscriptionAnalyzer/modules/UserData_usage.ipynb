```python
import matplotlib.pyplot as plt
import numpy as np

# Define screen time usage in hours for a month
screen_time = {
    "Netflix": 120,  # High Usage
    "Spotify": 50,   # Moderate Usage
    "Swiggy": 0,     # No Usage
    "Calls": 30,
    "WhatsApp": 90,
    "Instagram": 100
}

# Extracting data for visualization
apps = list(screen_time.keys())
usages = list(screen_time.values())

# Assigning colors based on usage levels
colors = ["red" if app == "Netflix" else "orange" if app == "Spotify" else "gray" if app == "Swiggy" else "blue" for app in apps]

# Create the histogram
plt.figure(figsize=(10, 6))
bars = plt.bar(apps, usages, color=colors)

# Add value labels on bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, yval, ha='center', fontsize=12)

# Customization
plt.xlabel("Apps")
plt.ylabel("Screen Time (Hours per Month)")
plt.title("Digital Wellbeing: Monthly Screen Time Usage")
plt.ylim(0, max(usages) + 20)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
